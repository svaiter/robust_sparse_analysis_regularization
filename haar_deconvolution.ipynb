Haar dictionary

```python
n = 128def haar_dict(n, scales=3, include_bf=False, normalization='unit'):    (Y,X) = mgrid[0:n,0:n]    I = mod(X-Y,n)    D = np.asarray([]).reshape((n,0))    for j in range(scales):        if normalization == 'tf':            rho = 2 ** (j+1)        elif normalization == 'unit':            rho = sqrt(2**(j+1))        d = zeros((n,1))        d[:2**j,:] = 1        #d[-1:-1-2**j+1,:]=-1        d[len(d)-2**j:,:]=-1        D = np.hstack((D,np.sum(d[I],axis=2)/rho))        if include_bf:            pass    return D
```

```python
D = haar_dict(n)Dtf = haar_dict(n, normalization='tf')Dtv = finite_diff_1d(n)imshow(D)
```

```python
def sig(n, type='up-down', eta=0.2):    x = zeros((n,1))    x[floor(n/2 - eta*n):floor(n/2 + eta*n),:] = 1    if type == 'up-up-down':        x[floor(n/2):floor(n/2 + eta*n),:] = 2    return x
```
