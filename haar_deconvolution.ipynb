```python
from __future__ import divisionimport numpy as npfrom scipy.linalg import toeplitzfrom rsar.criterions import criterions_dict, icfrom rsar.operators import finite_diff_1d
```

Haar dictionary definition

```python
n = 128def haar_dict(n, scales=3, normalization='unit'):    (X,Y) = mgrid[0:n,0:n]    I = mod(X-Y,n)    D = np.asarray([]).reshape((n,0))    for j in range(scales):        if normalization == 'tf':            rho = 2 ** (j+1)        elif normalization == 'unit':            rho = sqrt(2**(j+1))        else:            rho = 2**(normalization*(j+1))        d = zeros((n,1))        d[:2**j,:] = 1        #d[-1:-1-2**j+1,:]=-1        d[len(d)-2**j:,:]=-1        D = np.hstack((D,np.sum(d[I],axis=2)/rho))    return D
```

Gaussian filter

```python
def gaussian(n, sigma=0.5, shape=None):  sigma = max(abs(sigma), 1e-10)  if shape is None:    shape = max(int(6*sigma+0.5), 1)  x = numpy.arange(-(shape-1)/2.0, (shape-1)/2.0+1e-8)  Kx = numpy.exp(-x**2/(2*sigma**2)) / sqrt(2*pi*sigma*sigma)  op = toeplitz(np.vstack((Kx[:,np.newaxis], zeros((n-1,1)))), r=zeros((n,1)))  return op
```

Boxcar signal

```python
def sig(n, type='up-down', eta=0.2):    x = zeros((n,1))    x[floor(n/2 - eta*n):floor(n/2 + eta*n),:] = 1    if type == 'up-up-down':        x[floor(n/2):floor(n/2 + eta*n),:] = 2    return x
```

Evolution of $\textbf{IC}$ as a function of the standard deviation $\sigma$.

```python
n = 128D = haar_dict(n, scales=5, normalization='unit')Dtf = haar_dict(n, scales=5, normalization='tf')Dtv = finite_diff_1d(n)x0 = sig(n,eta=0.18, type='up-down')sigma_candidates = linspace(0.5, 3, num=50)res = zeros((sigma_candidates.shape[0], 3))for i, sigma in enumerate(sigma_candidates):    Phi = gaussian(n, sigma=sigma, shape=n)    res[i,0] = ic(D, Phi, x0, maxiter=100)    res[i,1] = ic(Dtf, Phi, x0, maxiter=100)    res[i,2] = ic(Dtv, Phi, x0, maxiter=100)
```
